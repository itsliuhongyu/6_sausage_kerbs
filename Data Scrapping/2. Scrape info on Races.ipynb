```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
import json
```

```python
with open('race_url_list.json', 'r') as f:
    url_list = json.load(f)
    
url_list
```

## Use panda to scrape results for each GP

```python
big_df = pd.DataFrame()  # initialize an empty dataframe

for url in url_list:
    time.sleep(3)  # wait for 3 seconds before each scrape
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    table = soup.find('table')

    if table is None:
        print(url + " is empty")
        continue

    df = pd.read_html(str(table))[0]
    race = url.split("/")[-3:-1]
    df['race'] = "/".join(race)
    big_df = pd.concat([big_df, df])
    print(url + " is scraped")

print("Scraping complete!")
```

```python

big_df
```

```python
big_df = big_df.drop(['Unnamed: 0', 'Unnamed: 8'], axis=1)
```

```python
big_df.to_csv('big_df.csv', index=False)
```

## Use beautifulsoup to append info in note

```python
notes_list = []

for url in url_list:
    response = requests.get(url)
    time.sleep(2)
    
    if response.status_code != 200:
        print(url + " is not accessible")
        continue
    
    soup = BeautifulSoup(response.content, 'html.parser')
    p_note = soup.find('p', {'class': 'note'})
    
    if p_note:
        note_text = p_note.text.strip()
        race = url.split("/")[-2] + "/" + url.split("/")[-3]
        year = url.split("/")[5]
        
        notes_dict = {"Note": note_text, "Race": race, "Year": year}
        notes_list.append(notes_dict)
        print(url + " is scraped")
        
    else:
        print(url + " is empty")
        continue
```

```python
notes_list
```

```python
df = pd.DataFrame(notes_list)
df.to_csv('notes_list.csv', index=False)
```

Note: 
1. The FIA marked 2020, 2021, and 2022 Emilia Romagna Grand Prix as Italy in its URL. These 3 races were held in Imola whilst other Italian GP were held in Monza.
2. 2020 Tuscani race is noted as Itally, but acrtually took place in Mugello.
3. 2020 Bahrain had two race with two different layout of the same track, I keep both notes as bahrain.
