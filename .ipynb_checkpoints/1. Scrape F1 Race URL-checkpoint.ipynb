```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
# Create a list of URLs for each year from 2012 to 2023
base_url = 'https://www.formula1.com/en/results.html/'
year_urls = [f'{base_url}{year}/races.html' for year in range(2012, 2024)]

# Initialize an empty list to store the race URLs
url_list = []

# Loop through the year URLs and extract the race URLs
for year_url in year_urls:
    # Send a GET request to the URL and get the HTML content
    response = requests.get(year_url)
    html_content = response.content

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the <select> element with name="meetingKey"
    meeting_select = soup.find('select', attrs={'name': 'meetingKey'})

    # Loop through the <option> elements and create a list of URLs
    for option in meeting_select.find_all('option'):
        if option['value'] != '':
            race_url = f'https://www.formula1.com/en/results.html/{year_url.split("/")[-2]}/races/{option["value"]}/race-result.html'
            url_list.append(race_url)

    # Print a message indicating that the year_url has been scraped
    print(year_url + " scraped")

    # Wait for 5 seconds before making the next request
    time.sleep(5)
```

```python
import json

# Save the url_list to a file
with open('race_url_list.json', 'w') as f:
    json.dump(url_list, f)
```
